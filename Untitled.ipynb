```python
for i, doc in enumerate(result_sorted):
        doc['rank'] = i + 1
for doc in result_sorted:
        data.append([doc['title'],doc['popularity'],doc['rank'],doc['spoken_languages']])
    
```

```python
grouped_data=[]
for item in data:
        title, popularity, rank, spoken_languages = item
        spoken_languages = spoken_languages.split(',')
        for language in spoken_languages:
            if language not in grouped_data:
                grouped_data[language] = []
            grouped_data[language].append([title, popularity, rank])
```
